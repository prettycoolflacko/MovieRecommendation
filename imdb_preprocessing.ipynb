```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

path = 'imdb_top_250_movies_with_ratings.csv'
df = pd.read_csv(path)

# Simpan salinan dataset asli untuk perbandingan nanti
df_awal = df.copy()

print("--- DATASET SEBELUM DIBERSIHKAN ---")
print("\nInfo Awal Dataset:")
df_awal.info()

print("\n5 Baris Pertama Data Awal:")
display(df_awal.head())
```

```python
print("\n--- Konversi Tipe Data ---")
print("Tipe data sebelum konversi:")
print(df.dtypes)

# Konversi kolom 'year' ke integer dan 'rating' ke float
df['year'] = df['year'].astype(int)
df['rating'] = df['rating'].astype(float)

# Duration awalnya dalam format string seperti '2h 30m', kita akan mengonversinya ke total menit sebagai integer
def convert_duration_to_minutes(duration_str):
    """Konversi format durasi 'Xh Ym' menjadi total menit."""
    if not isinstance(duration_str, str):
        return None
    
    hours = 0
    minutes = 0
    
    if 'h' in duration_str:
        parts = duration_str.split('h')
        hours = int(parts[0].strip())
        if len(parts) > 1 and 'm' in parts[1]:
            minutes = int(parts[1].replace('m', '').strip())
    elif 'm' in duration_str:
        minutes = int(duration_str.replace('m', '').strip())
    
    return (hours * 60) + minutes

# Buat kolom baru 'durasi_menit' dan hapus kolom 'duration' yang lama
df['durasi_menit'] = df['duration'].apply(convert_duration_to_minutes)
df = df.drop(columns=['duration'])

print("\nTipe data setelah konversi:")
print(df.dtypes)
print("\nHasil konversi durasi:")
display(df[['title', 'durasi_menit']].head())
```

```python
# Simpan dataset yang sudah dibersihkan ke file CSV baru
output_path = 'imdb_top_250_movies_cleaned.csv'
df.to_csv(output_path, index=False)
print(f"Dataset berhasil disimpan ke: {output_path}")

display(df.head())
```
